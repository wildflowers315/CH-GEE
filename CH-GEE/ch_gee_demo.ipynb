## Define Area of Interest

```python
Map = geemap.Map()
Map
```

```python
# Draw a polygon on the map
aoi = Map.draw_last_feature.geometry()
```

```python
# prompt: how to get area m2 of aoi (ee.geometry)

# Assuming 'aoi' is already defined as an ee.Geometry object.
area_m2 = aoi.area(maxError=1).getInfo() # Added maxError=1 to handle complex geometries
print(f"Area of AOI: {area_m2} m²")
area_ha = round(area_m2/10000)
print(area_ha)
polygon_area = ee.Number(area_ha)
print(polygon_area)
```

## Set Parameters

```python
# Set parameters
year = 2020
start_date = '01-01'
end_date = '12-31'
start_date_gedi = '2020-01-01'
end_date_gedi = '2022-12-31'
clouds_th = 0.5
quantile = 'rh98'
model = 'RF'
mask = 'FNF'
gedi_type = 'singleGEDI'

# Random Forest parameters
num_trees_rf = 500
var_split_rf = 3
min_leaf_pop_rf = 1
bag_frac_rf = 0.5
max_nodes_rf = None

# GBM parameters
num_trees_gbm = 100
shr_gbm = 0.1
sampling_rate_gbm = 0.8
max_nodes_gbm = None
loss_gbm = 'squared'

# CART parameters
max_nodes_cart = None
min_leaf_pop_cart = 1
```

## Run Canopy Height Mapping

```python
# Run canopy height mapping
classified = canopy_height_mapper(
    aoi=aoi,
    year=year,
    start_date=start_date,
    end_date=end_date,
    start_date_gedi=start_date_gedi,
    end_date_gedi=end_date_gedi,
    clouds_th=clouds_th,
    quantile=quantile,
    model=model,
    mask=mask,
    gedi_type=gedi_type,
    num_trees_rf=num_trees_rf,
    var_split_rf=var_split_rf,
    min_leaf_pop_rf=min_leaf_pop_rf,
    bag_frac_rf=bag_frac_rf,
    max_nodes_rf=max_nodes_rf,
    num_trees_gbm=num_trees_gbm,
    shr_gbm=shr_gbm,
    sampling_rate_gbm=sampling_rate_gbm,
    max_nodes_gbm=max_nodes_gbm,
    loss_gbm=loss_gbm,
    max_nodes_cart=max_nodes_cart,
    min_leaf_pop_cart=min_leaf_pop_cart
)
```
